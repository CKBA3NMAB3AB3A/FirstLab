```python
import pygame
import random

pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Анимация фигур")

FPS = 60
clock = pygame.time.Clock()

WHITE = (255, 255, 255)

class Shape:
    def __init__(self, shape_type, x, y, width, height, speed, color):
        self.type = shape_type
        self.x = x
        self.y = y
        self.width = width
        self.height = height
        self.speed = speed
        self.color = color
        self.direction = 1  # 1 - вправо, -1 - влево
        
        # Для треугольника
        self.points = [
            (x, y),
            (x - width//2, y + height),
            (x + width//2, y + height)
        ]

    def update(self):
        # Движение по горизонтали
        self.x += self.speed * self.direction
        
        # Обновление точек для треугольника
        if self.type == "triangle":
            self.points = [
                (self.x, self.y),
                (self.x - self.width//2, self.y + self.height),
                (self.x + self.width//2, self.y + self.height)
            ]
        
        # Проверка границ экрана
        if self.type == "circle":
            if self.x - self.width//2 <= 0 or self.x + self.width//2 >= WIDTH:
                self.direction *= -1
                self.color = (random.randint(0,255), random.randint(0,255), random.randint(0,255))
        else:
            if self.x <= 0 or self.x + self.width >= WIDTH:
                self.direction *= -1
                self.color = (random.randint(0,255), random.randint(0,255), random.randint(0,255))

    def draw(self, surface):
        if self.type == "square":
            pygame.draw.rect(surface, self.color, (self.x, self.y, self.width, self.height))
        elif self.type == "rectangle":
            pygame.draw.rect(surface, self.color, (self.x, self.y, self.width, self.height))
        elif self.type == "circle":
            pygame.draw.circle(surface, self.color, (self.x, self.y), self.width//2)
        elif self.type == "triangle":
            pygame.draw.polygon(surface, self.color, self.points)

    def is_clicked(self, pos):
        x, y = pos
        if self.type == "circle":
            distance = ((x - self.x)**2 + (y - self.y)**2)**0.5
            return distance <= self.width//2
        elif self.type == "triangle":
            # Проверка попадания в ограничивающий прямоугольник
            if (self.x - self.width//2 <= x <= self.x + self.width//2 and 
                self.y <= y <= self.y + self.height):
                # Более точная проверка с использованием маски
                mask = pygame.Surface((WIDTH, HEIGHT), pygame.SRCALPHA)
                pygame.draw.polygon(mask, (255,255,255), self.points)
                return mask.get_at((x, y))[3] > 0
            return False
        else:  
            return (self.x <= x <= self.x + self.width and 
                    self.y <= y <= self.y + self.height)

shapes = [
    Shape("square", 100, 300, 50, 50, 5, (255,0,0)),
    Shape("rectangle", 200, 250, 80, 100, 4, (0,255,0)),
    Shape("circle", 300, 300, 30, 0, 6, (0,0,255)),
    Shape("triangle", 400, 200, 60, 80, 3, (255,255,0))
]

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            mouse_pos = pygame.mouse.get_pos()
            for shape in shapes:
                if shape.is_clicked(mouse_pos):
                    shape.color = (random.randint(0,255), 
                                  random.randint(0,255), 
                                  random.randint(0,255))

    for shape in shapes:
        shape.update()

    screen.fill(WHITE)
    for shape in shapes:
        shape.draw(screen)
    
    pygame.display.flip()
    clock.tick(FPS)

pygame.quit()
```
