```python
import numpy as np
import scipy.stats as sts
```

```python
ni=np.array([146,97,73,34,23,10,6,3,4,2,2])
```

```python
n=sum(ni)
n
```

```python
lambda_hat=np.dot(xi,ni)/n
lambda_hat
```

```python
#P(X>=2)=1-P(X<2)=1-P(X=0)-P(X=1)
```

```python
P_A=1-(sts.poisson.pmf(0,lambda_hat)+sts.poisson.pmf(1,lambda_hat))
P_A
```

```python
#P(X>=2)=1-P(X<2)=1-P(X<=1)=1-P(X<=1)=1-F(X)
```

```python
P_A=1-sts.poisson.cdf(1,lambda_hat)
P_A
```

```python
xi=np.array([0.3,0.5,0.7,0.9,1.1,1.3,1.5,1.7,1.9,2.2,2.3])
ni=np.array([6,9,26,25,30,26,21,24,20,8,5])
```

```python
n=sum(ni)
```

```python
x_sr=np.dot(xi,ni)/n
x_sr
```
